```python
df=pd.read_csv('returns1020.csv')
df
```

```python
df["new_date"] = pd.to_datetime(df["Date"])
```

```python
df["new_date"]
```

```python
labels = ["S&P Ret", "FTSE Ret"] 
plt.figure(figsize = (20, 8))
plt.plot(df["new_date"], df["GSPCRet"],color="red")
plt.plot(df["new_date"], df["FTSERet"],color="blue")
plt.title("S&P vs FTSE Returns (2010 - 2020)", fontsize = 14, fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

```python
df_months = df[(df.new_date >= '2012-06-01') & (df.new_date <= '2012-12-31')]
```

```python
df_months
```

```python
plt.figure(figsize = (20, 8))
plt.plot(df_months["new_date"], df_months["GSPCRet"], color = "green")
plt.plot(df_months["new_date"], df_months["FTSERet"], color = "blue")
plt.title("S&P vs FTSE Returns for Year (2012)", fontsize = 14, fontweight = "bold")
plt.ylabel("Returns")
plt.xlabel("Date")
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

```python
#for the above chart:
#It show the returns for S&R Ret and FTSE Ret in year 2012 from From month five to month six
#And we choose this year because if we look at the chart that show returns from 2010 to 2020 ,we find that the changes of line in 2012 considered good 
```

```python
df=pd.read_csv('histogram_survey_data.csv')
df
```

```python
sns.set_style("white") 
plt.figure(figsize = (8, 6)) 
plt.hist(df["Age"], bins = 9,  color = "pink")
plt.title("Distribution of ages of participants for survey", fontsize = 14, weight = "bold")
plt.xlabel("Ages")
plt.ylabel("Income")
sns.despine() 
plt.show()
```

```python
#for tha above histogram
#we can found that the participants for survey from  the age of 25 to the age of 40 have higher income,So we find that the data is right skewed
```
